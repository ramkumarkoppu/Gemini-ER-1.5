````python
import google.generativeai as genai
import os
````

````python
# Configure the API key
genai.configure(api_key=os.environ.get("GOOGLE_API_KEY"))
````

````python
model = genai.GenerativeModel("gemini-2.5-pro")
chat = model.start_chat(history=[])
````

````python
chat.history
````

````python
while True:
    prompt = input("User: ")
    if prompt.lower() not in ["exit", "quit", "bye"]:
        response = chat.send_message(prompt)
        print(f"{chat.history[-1].role.capitalize()}: {chat.history[-1].parts[0].text}")
        print("\n" + "*" * 100 + "\n")
    else:
        print("Exiting the chat...")
        break
````

````python
instruction = '''
You are a coding expert specialized in front-end interfaces.
When I describe component, you will return the HTML with any CSS inline, 
necessary to build that component.
Do not give any explanations for the code. Return the code only.
'''
model = genai.GenerativeModel(model_name="gemini-2.5-pro", system_instruction=instruction)
````

````python
prompt = "A box with borders that contains YOUR_NAME in green aligned left and a list of links aligned right."
response = model.generate_content(prompt)
print(response.text)
````

````python
from IPython.display import HTML
HTML(response.text)
````

````python
html_code = response.text.strip().removeprefix('```html').removesuffix('```')
HTML(html_code)
````

````python
URL = "https://ia804501.us.archive.org/25/items/Winston_Churchill/1940-06-17_BBC_Winston_Churchill_Their_Finest_Hour.mp3"
````

````python
!wget -q $URL -O sample.mp3
````

````python
audio_file = genai.upload_file(path="sample.mp3")
prompt = "Create a transcript of the audio file."
model = genai.GenerativeModel(model_name="gemini-2.5-pro")
response = model.generate_content([prompt, audio_file])
print(response.text)
````

````python
Markdown(response.text)
````
