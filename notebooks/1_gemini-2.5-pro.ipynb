```python
import google.generativeai as genai
import os
```

```python
# Supply the API key
genai.configure(api_key=os.environ.get("$GOOGLE_API_KEY"))
```

```python
# Listing avaialble models
for model in genai.list_models():
    if "generateContent" in model.supported_generation_methods:
        print(model.name)
```

# Generate Text From Text Inputs

```python
model = genai.GenerativeModel("gemini-2.5-pro")
response = model.generate_content("Translate quantum machanics concepts into a story for teenagers.")
```

# Streaming

```python
prompt = "Write an imaginary scientific paper that proposes a new theory of quantum gravity."
```

```python
response = model.generate_content(prompt, stream=True)
for chunk in response:
    print(chunk.text, end="")
```

```python
generation_config = genai.types.GenerationConfig(
    candidate_count=1,
    stop_sequences=[';'],
    max_output_tokens=32000,
    temperature=0.9,
    top_p=1,
    top_k=1
)

model = genai.GenerativeModel("gemini-2.5-pro")
prompt = "Give me an example of a SQL SELECT statement."
response = model.generate_content(prompt, generation_config=generation_config)
print(response.text)
```
